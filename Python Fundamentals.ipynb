```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("sample_sales_data.csv")
df
```

```python
df["Revenue"] = df["Quantity"] * df["Price_per_Unit_INR"]
df
```

```python
print(np.corrcoef(df["Quantity"],df["Price_per_Unit_INR"] ))
```

```python
gb = df.groupby("Category").agg({"Quantity":"sum"})

plt.pie(gb["Quantity"], labels = gb.index)
plt.show()
```

```python
gb = df.groupby("Category").agg({"Revenue":"sum"})

plt.pie(gb["Revenue"], labels = gb.index)
plt.show()
```
